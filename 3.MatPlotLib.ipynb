### 예제:수직으로 던진 물체 

물체를 수직으로 던졌다고 가정한다.
이때 중력가속도는 9.8 m/s^2 이고 처음위치(y_)는 3 m, 처음속도(v_)는 3 m/s 라고하자.

가속도 a= -g이고

속도는 가속도를 적분한다.
v= v_ - g * t

위치는 한번더 적분하면 알수있다.
y= y_ + t*v_ + 0.5 * g * t^2

이것을 코드로 나타내면 아래와 같다.

```python
import matplotlib.pyplot as plt
import numpy as np

def throwBall():
    v_ = 5
    y_ = 3
    g = -9.8
    t = np.arange(0,1.5, 0.001)
    y= y_ + t*v_ + 0.5 * g * np.power(t,2)
    plt.title("throwBall")
    plt.xlabel("Time")
    plt.ylabel("y")
    plt.plot(t, y)
    plt.show()
```

하나의 함수로 위의 문제를 정의했으며, 함수를 실행 시키면 그래프가 나타날 것이다.

np.power(t,2)는 t의 제곱을 코드로 표현한것이다.

```python
throwBall()
```

물체는 올라갔다가 다시 떨어지는것을 그래프로 알 수 있다.

### 예제: 자유낙하
자유낙하의 경우를 생각해보는 것은 매우 쉬운 일이다.

단, 이때 그래프 3개를 나눠서 한번에 볼 수 있게 그려보자


```python
t=np.linspace(0,3,1000)
g=np.array([10])+0*t
v=15-g*t
x=15*t-1/2*g*t**2


a=plt.figure().add_subplot(3,1,1)
a.plot(t,g,'g')
plt.ylabel('Acceleration')
V=plt.figure().add_subplot(3,1,2)
V.plot(t,v,'b')
plt.ylabel('Velocity')
p=plt.figure().add_subplot(3,1,3)
p.plot(t,x,'r')
plt.ylabel('Positon')
plt.xlabel('Time')

plt.show()
```

### 예제:진동 
우리는 감쇠 진동과 조화진동의 그래프를 그려볼 것이다.

단, 나중에 연립방정식을 풀때는 유도할 것이지만 이번에는 답을 안다고 가정하고 풀어보자


```python
time = np.linspace(0.0,10.0,100)                        
damped=np.exp(-time/3)*np.sin(time*3)
undamped=0.3*np.sin(time*3)
plt.plot(time,damped,'m-^')                             
plt.plot(time,undamped,'g-')                  
plt.legend(['damped','constant amplitude'],loc='upper right')   
plt.xlabel('Time(s)') 
plt.show()
```
