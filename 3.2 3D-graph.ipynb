## 3D Graph 그리기

파이썬 버전 3.6.0


3차원에서 물체또는 입자의 운동을 기술할려면 3D 그래프가 필요 할 수도 있다.
 
마찬가지로 matplotlib를 이용할 것이지만 mpl_toolkits.mplot3d 에서 Axes3D라는 3D 그래프를 그리는데 필요한 함수를 가져와야한다.

마치 numpy에서 선형대수적 문제를 풀때 numpy.linalg에서 모두 가져오는 것 처럼 말이다

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

그래프를 그릴때 수식을 쉽게 쓰거나 다른 값들을 쉽게 얻어오기 위해 numpy도 불러왔다.

예시로 그래프를 그리고 알아가자

### 3.2.1 선으로된 그래프 그리기:parametric curve

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z, label='parametric curve')
ax.legend()

plt.show()
```

parametric curve, 3D 그래프가 생겼다. 

일단 딱 봐도 알 수 있는것부터 간단하게 알아보자

theta는 -4파이부터 +4파이까지, z는 -2부터 2까지 정의된다.

r은 z^2+1로 정의하고, 이 r로 다시 x와 y를 정의했다.

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
```

이것은 우리가 그릴 ax 그래프가 3차원이라고 설정하는것이다.


```python
ax.plot(x, y, z, label='parametric curve')
ax.legend()
```

따라서 ax 그래프에는 축이 3개 필요하고 x축에는 x, y축에는 y, z축에는 z를 넣어준다. 

(순서대로 x,y,z 축에 들어갈 함수를 넣어주면 된다)

label은 그래프의 이름을 넣어준다.

이제 plt.show()로 그려주면 끝이다.

### 3.2.2 산란 그래프
그래프가 선이 아니라, 점들이 퍼져있는 그래프를 그리고 싶다면 어떻게 그려야 할까?

일단 점들이 원하는 지점 사이에서 랜덤한 위치에 퍼진다고 가정하자.

라이브러리를 불러오는 것은 위해서 했으므로 생략하자.

그럴려면 점들은 x,y,z축의 좌표를 가져야한다.  다른의미로는 x,y,z축은 가진 점의 갯수가 같아야한다.

따라서 아래 함수로 나타낼 수 있다.


```python
def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin
```

이제 뭘해야될까? 위에서 3.2.1에서 본대로 우리가 3D 그래프를 그릴꺼라고 설정하자.


```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

그래프를 그릴려면 각 점의 정보가 필요하다.

```python
n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m)
```

점은 100개 그린다고 가정하자

for문이 돌면서 x는 23부터 32로 랜덤하게 100개만들고 배열 xs에 넣은것이다.
마찬가지로 ys도 똑같다.

z는 특별한데 -50부터 -25인 경우와 -30부터 -5인경우로 나눠져있다.

scatter함수를 보면 c=c가 점의 색을 정하고, marker=m이 점의 모양을 정한다.

for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)] 은

for문에서 z가 -50부터-25인것은 빨간색 동그라미, -30부터 -5인것은 파란색 삼각형이 된다.

```python
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

각축의 이름을 정해주고 plt.show()로 그려주면 된다.
