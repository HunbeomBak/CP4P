### Pandas로 데이터 불러와서 numpy로 모델 만들기

y= W * X + b 로 나타낼 것이다.

pandas 호출

```python
import pandas as pd
```

데이터 불러오기

csv 타입 데이터 로드, separate는 빈공간으로 지정하고, Column은 없음

ML_1.Pandas 참고

```python
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data' #Data URL
df_data = pd.read_csv(data_url, sep='\s+', header = None) 
df_data.columns = ['CRIM','ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO' ,'B', 'LSTAT', 'MEDV'] 
df_data.head()
```

상수항(W=y= W * X + b에서 b)은 weight가 1이다.  

weight 0의 값 추가해야 한다.

```python
df_data['weight_0'] = 1
```

MEDV는 y값이므로 모델을 표현할때 없어도된다.

```python
df_data= df_data.drop("MEDV", axis=1)
df_data.head()
```

이제 위 표를 matrix로 변환한다.

아직 weight는 학습시키지 않았으므로 랜덤값을 사용한다.

```python
df_matrix = df_data.as_matrix() 
weight_vector =  np.random.random_sample((14, 1)
```

df_matrix가 X이며 weight_vector가 W이다.

이제 곱하면 모델이 완성된다.

```python
df_matrix.dot(weight_vector)
```

```python
df_matrix.dot(weight_vector).shape
```
