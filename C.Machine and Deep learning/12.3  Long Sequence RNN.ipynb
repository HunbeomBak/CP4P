```python
import tensorflow as tf
import numpy as np
```

좀 더 긴 문장을 가르쳐 보자

그럴려면 또 우리가 입력할 데이터를 인덱싱하고 원핫 인코딩을 해줘야한다.

문자열 데이터를 자동으로 인덱싱 처리해보자

```python
sample = "if you want you"
```

sample에서 사용된 문자만 뽑아올려면 List의 set을 이용하면 된다

```python
idx2char=list(set(sample))

print(idx2char)
```

이제 쓰인 문자마다 번호를 매겨주자

```python
char2idx = {c: i for i, c in enumerate(idx2char)}

print(char2idx)
```

총 0부터 9까지의 인덱스가 생겼다. 

원핫인코딩을 할때 num_classes = 10이다

우리가 줄 smaple="if you want you" 를 번호로 바꾸어보자

```python
sample_idx=[char2idx[c]for c in sample]

print(sample_idx)
```

앞장에서는 필요한 변수 (sequence_length, num_classes)를 일일이 새서 넣어줬지만 자동적으로 결정할 수 있다.


```python
dic_size = len(char2idx) #RNN input size (= one hot size)
rnn_hidden_size = len(char2idx) # RNN output size
num_classes = len(char2idx) #final output size (RNN or softmax, etc)

batch_size = 1 #sample이 하나이므로 한개

squence_length = len(char2idx) -1 
```

이제 x,y 데이터를 정해주자.

우리는 if you want yo를 입력해서 f you want you를 나오게 할 것이다

```python
x_data =[sample_idx[:-1]]
y_data =[sample_idx[1:]]

X= tf.placeholder(tf.int32,[None, sequence_length])
Y= tf.placeholder(tf.int32,[None, sequence_length])
```

입력을 원핫으로 주어야 한다.

일일이 바꿀 걱정하지 말고 텐서플로우의 함수를 이용하자.

```python
X_one_hot = tf.one_hot(X, num_classes)
```

이제 cell을 만들어 보자

```python
cell = tf.contrib.rnn.BasicLSTMCell(num_units=rnn_hidden_size, state_is_tuple=True)
initial_state=cell.zero_state(batch_size, tf.float32)
outputs, _state=tf.nn.dynamic_rnn(cell,X_one_hot,initial_state=initial_state,dtype=tf.float32)
```

cell을 만들었다면 이제 얼마나 정확한지 loss함수를 만들어 확인해보자

```python
weight= tf.ones([batch_size,sequence_length])

sequence_loss=tf.contrib.seq2seq.sequence_loss(logits=outputs,targets=Y,weights=weight)
loss =tf.reduce_mean(sequence_loss)
train=tf.train.AdamOptimizer(learning_rate=0.1).minimize(loss)

prediction = tf.argmax(outputs, axis=2)
```

학습 가즈아ㅏㅏㅏㅏㅏㅏ

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(3000):
        l,_=sess.run([loss,train],feed_dict={X:x_data, Y: y_data})
        result=sess.run(prediction,feed_dict={X:x_data})
        
        result_str= [idx2char[c]for c in np.squeeze(result)]
        print(i,"loss:",l,"Prediction:",''.join(result_str))
```
