## Series selection

컬럼에서 특정 시리즈만 뽑을 수 있다.

```python
account_serires = df["account"]
account_serires
```

```python
account_serires[:3]
```

원하는 번호만 뽑고 싶을때

```python
account_serires[[1,5,2]]
```

조건을 넣어주면 true인 애들만 뽑을 수 있다.

```python
account_serires[account_serires<250000]
```

## index 변경

인덱스를 단순히 번호가 아니라 직접 지정해줄 수 있다,

```python
df.index = df["account"]
```

```python
df.head()
```

인덱스와 series에 account가 겹침으로 지워주자

```python
del df["account"]
df.head()
```

```python
df[["name","street"]][:2]
```

loc와 iloc으로 쉽게 데이터를 뽑아 올 수 있다.

loc은index와 원하는 column을 str로 넣어주면 뽑아온다,

```python
df.loc[[211829,320563],["name","street"]]
```

iloc은 index의 넘버를 기준으로 뽑아온다.

```python
df[["name", "street"]].iloc[:10]
```
